```python
from pandas import read_csv
```

```python
url = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-01-21/spotify_songs.csv'
dataset = read_csv(url, index_col=0)
dataset.shape
```

```python
dataset.head()
```

```python
dataset = dataset.dropna()
dataset
```

```python
dataset = dataset.drop(['playlist_id','playlist_name','track_album_id','playlist_subgenre'], axis=1)
dataset
```

```python
duplicates_list = dataset.duplicated(keep='first')
duplicates_list
```

```python
datasetwd = dataset[~duplicates_list] #dataset without duplicates
datasetwd
```

```python
datasetwd.groupby('playlist_genre').count()
```

```python
datasetwd.groupby('playlist_genre').size()
```
